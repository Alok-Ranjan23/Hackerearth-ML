### Drop useless columns from training Dataset, Need to drop from Test Dataset too.
 * customer_id - Represents the unique identification number of a customer
 * Name - Represents the name of a customer
 * security_no - Represents a unique security number that is used to identify a person
 * referral_id - Represents a referral ID

```python
train = train.drop('customer_id',axis=1)
train = train.drop('Name',axis=1)
train = train.drop('security_no',axis=1)
train = train.drop('referral_id',axis=1)
```

```python
train.info()
```

### DataFields with NA entries:: 
* region_category - Represents the region that a customer belongs to
* preferred_offer_types - Represents the type of offer that a customer prefers 
* points_in_wallet - Represents the points awarded to a customer on each transaction

```python
train['region_category'].value_counts()
```

```python
train = train[train['region_category'].notna()]
```

```python
train.info()
```

```python
train['preferred_offer_types'].value_counts()
```

```python
train = train[train['preferred_offer_types'].notna()]
```

```python
train.info()
```

```python
train['points_in_wallet'] = train['points_in_wallet'].fillna(round(train['points_in_wallet'].median(),2))
```
